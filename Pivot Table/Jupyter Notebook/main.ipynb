```python
import warnings
warnings.filterwarnings("ignore")

import gspread
from oauth2client.service_account import ServiceAccountCredentials


scope = ["https://spreadsheets.google.com/feeds"]
credentials = ServiceAccountCredentials.from_json_keyfile_name("total-market-335122-1b57a33cd1c2.json",scope)

gc = gspread.authorize(credentials)
```

```python
import pandas as pd

from IPython.display import display
```

```python
worksheet = gc.open_by_key("1gqqbcAY7eDmEqvvubvnX7mezSAHV-cjL7sNHf3VTboc").worksheet("Sheet1")
row = worksheet.get_all_records()
```

```python
dataset = pd.DataFrame(row)
display(dataset)
```

```python
df_pivot_table = pd.pivot_table(dataset, index="Motor", columns="Area", values="Penjualan", aggfunc="sum", margins=True, margins_name="Grand Total")
display(df_pivot_table)
```

```python
df_pivot_table = pd.pivot_table(dataset, index=["Motor","Bulan"], columns="Area", values="Penjualan", aggfunc="sum", margins=True, margins_name="Grand Total")
display(df_pivot_table)
```

```python
df_pivot_table = pd.pivot_table(dataset, index=["Motor"], columns=["Area","Bulan"], values="Penjualan", aggfunc="sum", margins=True, margins_name="Grand Total")
display(df_pivot_table)
```
