```python
import pandas as pd

from sklearn.preprocessing import OneHotEncoder ,OrdinalEncoder, LabelEncoder
from IPython.display import display
```

```python
data = {
        "kota" : ["Jakarta", "Bandung", "Yogyakarta", "Surabaya", "Malang"],
        "tingkat_populasi" : ["Tinggi", "Sedang", "Rendah", "Sedang", "Sedang"]
       }

dataset = pd.DataFrame(data)
display(dataset)
```


## Encoding Data Kategorikal Nominal -> Kota -> get_dummies

```python
nominal_kota = pd.get_dummies(dataset["kota"])
display(nominal_kota)
```

```python
new_dataset = pd.merge(dataset, nominal_kota, left_index=True, right_index=True)
display(new_dataset)
```

```python
onehotencoder = OneHotEncoder(sparse=False)
nominal_kota = onehotencoder.fit_transform(dataset[["kota"]])
column_nominal_kota = onehotencoder.categories_
df_nominal_kota = pd.DataFrame(nominal_kota, columns=column_nominal_kota[0])
display(df_nominal_kota.columns)
```

```python
new_dataset = pd.merge(dataset, df_nominal_kota, left_index=True, right_index=True)
new_dataset = new_dataset.drop(["kota"], axis=1)
display(new_dataset)
```

## Encoding data Kategorikal Ordinal -> kota - Label Encoder & OrdinalEncoder

```python
LE = LabelEncoder()
new_dataset["populasi"]= LE.fit_transform(new_dataset["tingkat_populasi"])
display(new_dataset)
```
