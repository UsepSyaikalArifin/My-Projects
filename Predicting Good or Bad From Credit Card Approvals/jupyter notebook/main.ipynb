```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns

from IPython.display import display
```

```python
applications = pd.read_csv(r"D:\Usep Syaikal Arifin\Computer Programming\Python\MyProjects\Predicting Good or Bad From Credit Card Approvals\dataset\application_record.csv")

print("Shape of dataset = %d Baris & %d Kolom"%applications.shape)
print("Informasi dataset : ")
display(applications.info())
print("Mengecek Missing Value dataset : ")
display(applications.isnull().sum())
display(applications.head())
```

```python
def data_cleansing(data):
    data = data.rename(columns={
                        "CODE_GENDER" : "jenis kelamin",
                        "FLAG_OWN_CAR" : "punya mobil ?",
                        "FLAG_OWN_REALTY" : "punya properti ?",
                        "CNT_CHILDREN" : "jumlah anak",
                        "AMT_INCOME_TOTAL" : "total penghasilan",
                        "NAME_INCOME_TYPE" : "asal penghasilan",
                        "NAME_EDUCATION_TYPE" : "jenis pendidikan",
                        "NAME_FAMILY_STATUS" : "status pernikahan",
                        "NAME_HOUSING_TYPE" : "jenis tempat tinggal",
                        "DAYS_BIRTH" : "usia",
                        "DAYS_EMPLOYED" : "tanggal daftar",
                        "FLAG_MOBIL" :"punya mobil phone ?",
                        "FLAG_WORK_PHONE" : "punya tlp rumah ?",
                        "FLAG_PHONE" : "punya hp ?",
                        "FLAG_EMAIL" : "punya email ?",
                        "OCCUPATION_TYPE" : "jenis pekerjaan",
                        "CNT_FAM_MEMBERS" : "jumlah anggota keluarga"
                      })

    data["jumlah anggota keluarga"] = (data["jumlah anak"] + data["jumlah anggota keluarga"]).astype("int64")
    data["total penghasilan"] = data["total penghasilan"].astype("int64")
    
    data = data.drop(["jumlah anak","punya mobil phone ?","punya mobil phone ?","punya tlp rumah ?","punya hp ?","punya email ?","jenis pekerjaan"],axis=1)

    data["usia"] = (abs(round(data["usia"]/365))).astype("int64")
    data["tanggal daftar"] = (round(data["tanggal daftar"]/365)).astype("int64")

    data = data[["ID","jenis kelamin","usia","jenis pendidikan","jenis tempat tinggal","jumlah anggota keluarga","status pernikahan","total penghasilan","asal penghasilan","punya mobil ?","punya properti ?","tanggal daftar"]]

    jenis_kelamin = {
                      "M" : "Laki-Laki",
                      "F" : "Perempuan"
                    }
    data["jenis kelamin"] = data["jenis kelamin"].map(jenis_kelamin)

    jenis_pendidikan = {
                        "Secondary / secondary special":"SMP",
                        "Lower secondary":"SD",
                        "Higher education":"SMA",
                        "Incomplete higher":"SMA",
                        "Academic degree":"Kuliah"
                       }
    data["jenis pendidikan"] = data["jenis pendidikan"].map(jenis_pendidikan)

    jenis_tempat_tinggal = {
                            "House / apartment" : "Apartement",
                            "With parents": "Tinggal Bersama Keluarga",
                            "Municipal apartment" : "Apartement",
                            "Rented apartment": "Apartement",
                            "Office apartment": "Apartement",
                            "Co-op apartment": "Apartement"
                            }
    data["jenis tempat tinggal"] = data["jenis tempat tinggal"].map(jenis_tempat_tinggal)

    family_status = {
                     "Single / not married":"Sendiri",
                     "Separated":"Sendiri",
                     "Widow":"Sendiri",
                     "Civil marriage":"Menikah",
                     "Married":"Menikah"
                    }
    data["status pernikahan"] = data["status pernikahan"].map(family_status)

    income_type = {
                  "Commercial associate":"Bekerja",
                  "State servant":"Bekerja",
                  "Working":"Bekerja",
                  "Pensioner":"Pensiunan",
                  "Student":"Pelajar"
                  }
    data["asal penghasilan"] = data["asal penghasilan"].map(income_type)

    kepemilikan_mobil = {
                          "Y" : "Punya Mobil",
                          "N" : "Tidak Punya Mobil"
                        }
    data["punya mobil ?"] = data["punya mobil ?"].map(kepemilikan_mobil)

    kepemilikan_property = {
                            "Y" : "Punya Property",
                            "N" : "Tidal Punya Property"
                           }
    data["punya properti ?"] = data["punya properti ?"].map(kepemilikan_property)
    
    return data

idn_applications = data_cleansing(applications)

print("Informasi idn_applications : ")
display(idn_applications.info())
display(idn_applications.head())
```

```python
credits = pd.read_csv(r"D:\Usep Syaikal Arifin\Computer Programming\Python\MyProjects\Predicting Good or Bad From Credit Card Approvals\dataset\credit_record.csv")

print("Shape of dataset = %d Baris & %d Kolom"%credits.shape)
print("Informasi dataset : ")
display(credits.info())
print("Mengecek Missing Value dataset : ")
display(credits.isnull().sum())
display(credits.head())
```

```python
def ubah_nama_column_table_credits(data):
    dict_columns = {
                    "MONTHS_BALANCE" : "tagihan bulan ke",
                    "STATUS" : "status pembayaran tagihan"
                   }
    data = data.rename(dict_columns, axis=1)

    return data

idn_credits = ubah_nama_column_table_credits(credits)

display(idn_credits.head())
```

```python
color = sns.color_palette("bright")
fig, axs = plt.subplots(1, 1, figsize=(7, 5))
sns.countplot(data=idn_credits, x="status pembayaran tagihan", order=idn_credits["status pembayaran tagihan"].value_counts().index, palette=color, ax=axs)
axs.set_title("Komposisi Status Pembayaran Tagihan", fontsize=20, fontweight="bold", loc="left", pad=20)
axs.set_xlabel("Status Pembayaran Tagihan", fontsize=15)
axs.yaxis.set_visible(False)
for p in axs.patches:
    height = p.get_height()
    width = p.get_width()
    x = p.get_x()
    y = p.get_y()
    axs.annotate(
                  height,
                  xy=(x + width/2, height),
                  ha = "center",
                  va = "center",
                  xytext = (0, 10),
                  textcoords= "offset points"
                )
sns.despine(left=True, top=True, right=True)
plt.tight_layout()
plt.show()
```

```python
def pivot_idn_credits(data):
    pivot_idn_credits = data.pivot(index="ID", columns="tagihan bulan ke", values="status pembayaran tagihan")

    pivot_idn_credits["awal bulan peminjaman"] = data.groupby(["ID"])["tagihan bulan ke"].min()
    pivot_idn_credits["akhir bulan peminjaman"] = data.groupby(["ID"])["tagihan bulan ke"].max()
    pivot_idn_credits["dilunasi dalam x bulan"] = pivot_idn_credits["akhir bulan peminjaman"] - pivot_idn_credits["awal bulan peminjaman"]
    pivot_idn_credits["dilunasi dalam x bulan"] += 1

    pivot_idn_credits["mampu melunasi tagihan x kali"] = pivot_idn_credits[pivot_idn_credits.iloc[:,0:61] == "C" ].count(axis=1)

    pivot_idn_credits["tagihan telat 1-29 hari x kali"] = pivot_idn_credits[pivot_idn_credits.iloc[:,0:61] == "O" ].count(axis=1)
    pivot_idn_credits["tagihan telat 30-59 hari x kali"] = pivot_idn_credits[pivot_idn_credits.iloc[:,0:61] == "1" ].count(axis=1)
    pivot_idn_credits["tagihan telat 60-89 hari x kali"] = pivot_idn_credits[pivot_idn_credits.iloc[:,0:61] == "2" ].count(axis=1)
    pivot_idn_credits["tagihan telat 90-119 hari x kali"] = pivot_idn_credits[pivot_idn_credits.iloc[:,0:61] == "3" ].count(axis=1)
    pivot_idn_credits["tagihan telat 120-149 hari x kali"] = pivot_idn_credits[pivot_idn_credits.iloc[:,0:61] == "4" ].count(axis=1)
    pivot_idn_credits["tagihan telat > 150 hari x kali"] = pivot_idn_credits[pivot_idn_credits.iloc[:,0:61] == "5" ].count(axis=1)

    pivot_idn_credits["total telat bayar"] = pivot_idn_credits["tagihan telat 1-29 hari x kali"] + pivot_idn_credits["tagihan telat 30-59 hari x kali"] + pivot_idn_credits["tagihan telat 60-89 hari x kali"] +  pivot_idn_credits["tagihan telat 90-119 hari x kali"] + pivot_idn_credits["tagihan telat 120-149 hari x kali"] + pivot_idn_credits["tagihan telat > 150 hari x kali"]


    pivot_idn_credits["tidak ada pinjaman x kali"] = pivot_idn_credits[pivot_idn_credits.iloc[:,0:61] == "X" ].count(axis=1)

    return pivot_idn_credits

pivot_table_credits_clean = pivot_idn_credits(idn_credits)
# display(pivot_table_credits_clean.iloc[50:61,61:74])
display()
```

```python
def membuat_target(data):
    status_pelanggan = []
    for i, baris in data.iterrows():
        mampu_bayar = baris["mampu melunasi tagihan x kali"]
        total_telat_bayar = baris["total telat bayar"]
        tidak_meminjam = baris["tidak ada pinjaman x kali"]

        if total_telat_bayar == 0:
            if mampu_bayar >= tidak_meminjam or mampu_bayar <= tidak_meminjam:
                status_pelanggan.append(1)
            elif mampu_bayar == 0 and tidak_meminjam == 1:
                status_pelanggan.append(1)
        elif total_telat_bayar != 0:
            if total_telat_bayar <= mampu_bayar:
                status_pelanggan.append(1)
            elif total_telat_bayar > mampu_bayar:
                status_pelanggan.append(0)
        elif mampu_bayar == 0 and tidak_meminjam != 0:
            if total_telat_bayar >= tidak_meminjam or total_telat_bayar <= tidak_meminjam:
                status_pelanggan.append(0)
        else:
            status_pelanggan.append(1)

    return status_pelanggan
status_pelanggan = membuat_target(pivot_table_credits_clean)
display(status_pelanggan[0:5])
```

```python
def membuat_df_target(data):
    df_target = pd.DataFrame()
    df_target["ID"] = data.index
    df_target["Status"] = status_pelanggan

    return df_target

df_target = membuat_df_target(pivot_table_credits_clean)
display(df_target["Status"].value_counts())
```

```python
dataset_clean = pd.merge(idn_applications, df_target, on="ID", how="inner")
display(dataset_clean.head())
```
