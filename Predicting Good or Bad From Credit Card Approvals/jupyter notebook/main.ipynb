```python
credits = pd.read_csv(r"D:\Usep Syaikal Arifin\Computer Programming\Python\MyProjects\Predicting Good or Bad From Credit Card Approvals\dataset\credit_record.csv")

print("Shape of dataset = %d Baris & %d Kolom"%credits.shape)
print("Informasi dataset : ")
display(credits.info())
print("Mengecek Missing Value dataset : ")
display(credits.isnull().sum())
display(credits.head())
```

```python
def ubah_nama_column_table_credits(data):
    dict_columns = {
                    "MONTHS_BALANCE" : "tagihan bulan ke",
                    "STATUS" : "status pembayaran tagihan"
                   }
    data = data.rename(dict_columns, axis=1)

    return data

idn_credits = ubah_nama_column_table_credits(credits)

display(idn_credits.head())
```

```python
color = sns.color_palette("bright")
fig, axs = plt.subplots(1, 1, figsize=(7, 5))
sns.countplot(data=idn_credits, x="status pembayaran tagihan", order=idn_credits["status pembayaran tagihan"].value_counts().index, palette=color, ax=axs)
axs.set_title("Komposisi Status Pembayaran Tagihan", fontsize=20, fontweight="bold", loc="left", pad=20)
axs.set_xlabel("Status Pembayaran Tagihan", fontsize=15)
axs.yaxis.set_visible(False)
for p in axs.patches:
    height = p.get_height()
    width = p.get_width()
    x = p.get_x()
    y = p.get_y()
    axs.annotate(
                  height,
                  xy=(x + width/2, height),
                  ha = "center",
                  va = "center",
                  xytext = (0, 10),
                  textcoords= "offset points"
                )
sns.despine(left=True, top=True, right=True)
plt.tight_layout()
plt.show()
```

```python
def pivot_idn_credits(data):
    pivot_idn_credits = data.pivot(index="ID", columns="tagihan bulan ke", values="status pembayaran tagihan")

    pivot_idn_credits["awal bulan peminjaman"] = data.groupby(["ID"])["tagihan bulan ke"].min()
    pivot_idn_credits["akhir bulan peminjaman"] = data.groupby(["ID"])["tagihan bulan ke"].max()
    pivot_idn_credits["dilunasi dalam x bulan"] = pivot_idn_credits["akhir bulan peminjaman"] - pivot_idn_credits["awal bulan peminjaman"]
    pivot_idn_credits["dilunasi dalam x bulan"] += 1

    pivot_idn_credits["tagihan dilunasi x kali"] = pivot_idn_credits[pivot_idn_credits.iloc[:,0:61] == "C" ].count(axis=1)

    pivot_idn_credits["tagihan telat 1-29 hari x kali"] = pivot_idn_credits[pivot_idn_credits.iloc[:,0:61] == "O" ].count(axis=1)
    pivot_idn_credits["tagihan telat 30-59 hari x kali"] = pivot_idn_credits[pivot_idn_credits.iloc[:,0:61] == "1" ].count(axis=1)
    pivot_idn_credits["tagihan telat 60-89 hari x kali"] = pivot_idn_credits[pivot_idn_credits.iloc[:,0:61] == "2" ].count(axis=1)
    pivot_idn_credits["tagihan telat 90-119 hari x kali"] = pivot_idn_credits[pivot_idn_credits.iloc[:,0:61] == "3" ].count(axis=1)
    pivot_idn_credits["tagihan telat 120-149 hari x kali"] = pivot_idn_credits[pivot_idn_credits.iloc[:,0:61] == "4" ].count(axis=1)
    pivot_idn_credits["tagihan telat > 150 hari x kali"] = pivot_idn_credits[pivot_idn_credits.iloc[:,0:61] == "5" ].count(axis=1)

    pivot_idn_credits[" total telat bayar"] = pivot_idn_credits["tagihan telat 1-29 hari x kali"] + pivot_idn_credits["tagihan telat 30-59 hari x kali"] + pivot_idn_credits["tagihan telat 60-89 hari x kali"] +  pivot_idn_credits["tagihan telat 90-119 hari x kali"] + pivot_idn_credits["tagihan telat 120-149 hari x kali"] + pivot_idn_credits["tagihan telat > 150 hari x kali"]


    pivot_idn_credits["tidak ada pinjaman x kali"] = pivot_idn_credits[pivot_idn_credits.iloc[:,0:61] == "X" ].count(axis=1)

    return pivot_idn_credits

pivot_table_credits_clean = pivot_idn_credits(idn_credits)
display(pivot_table_credits_clean.tail(30))
```
